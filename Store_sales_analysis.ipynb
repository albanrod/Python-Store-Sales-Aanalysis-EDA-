```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#rename column
df.rename(columns= {'Marital_Status':'Married'})
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc) for numeric columns
df.describe()
```

```python
# use describe() for specific columns as user id,marital status has no significance for numeric analysis
df[['Age', 'Orders', 'Amount']].describe()
```

# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
   ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
print(sales_gen)
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
print(sales_age)
plt.figure(figsize=(10, 6))
ax=sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Gender')
plt.title('Total Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

### State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data = sales_state, x = 'State',y= 'Orders')
plt.title('Total Orders from top 10 states')
plt.xlabel('State')
plt.ylabel('Orders')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,5)})
ax=sns.barplot(data = sales_state, x = 'State',y= 'Amount')
plt.title('Total amount/sales from top 10 states')
plt.xlabel('State')
plt.ylabel('Amount')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(6,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_married = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
print(sales_married)
sns.set(rc={'figure.figsize':(6,5)})
ax=sns.barplot(data = sales_married, x = 'Marital_Status',y= 'Amount', hue='Gender')
plt.title('Total sales as per Marital_Status')
plt.xlabel('Marital_Status')
plt.ylabel('Amount')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
print(sales_occupation)
```

```python
sns.set(rc={'figure.figsize': (8, 4)})
ax=sns.barplot(data=sales_occupation, y='Occupation', x='Amount', orient='h')
plt.xlabel('Amount')
plt.ylabel('Occupation')
plt.title('Total Sales by Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
print(sales_product_category)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_product_category, x = 'Product_Category',y= 'Amount')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_product_id = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
print(sales_product_id)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_product_id, x = 'Orders',y= 'Product_ID', orient='h')
```

```python
# Grouping and calculating total orders by product ID
sales_product_id = df.groupby(['Product_ID', 'Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
print(sales_product_id)
# Create the bar plot
sns.set(rc={'figure.figsize': (10, 6)})
ax = sns.barplot(data=sales_product_id, x='Product_ID', y='Orders', hue='Product_Category')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.xlabel('Product ID')
plt.ylabel('Total Orders')
plt.title('Top 10 Products by Total Orders')
plt.legend(title='Product Category')
plt.show()
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
